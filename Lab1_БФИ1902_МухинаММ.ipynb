## Задание №3
Описание условия задания:  
Реализовать методы сортировки строк числовой матрицы в соответствии с заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со временем стандартной функции сортировки. Испытания проводить на сгенерированных матрицах  
Методы:  
| Выбором | Вставкой | Обменом | Шелла | Турнирная | Быстрая сортировка | Пирамидальная | 

```python
import random
import copy
import datetime


def test_sort(matr1, func, namesort):
    start_time = datetime.datetime.now()
    for i in range(len(matr1)):
        func(matr1[i])
    end_time = datetime.datetime.now()
    print(namesort)
    print(end_time - start_time)
    for i in range(len(matr1) - 1):
        print(matr1[i])


# Выбором
def selection_sort(alist):
    for i in range(0, len(alist) - 1):
        smallest = i
        for j in range(i + 1, len(alist)):
            if alist[j] < alist[smallest]:
                smallest = j
        alist[i], alist[smallest] = alist[smallest], alist[i]


# Вставкой
def insertion_sort(alist):
    for i in range(1, len(alist)):
        temp = alist[i]
        j = i - 1
        while j >= 0 and temp < alist[j]:  # пока j не отрицательное и temp меньше, чем элемент под индексом j
            alist[j + 1] = alist[j]
            j = j - 1
        alist[j + 1] = temp


# Обменом
def bubble_sort(alist):
    for i in range(len(alist) - 1, 0, -1):  # от (длины списка - 1) до 1
        no_swap = True
        for j in range(0, i):  # если значение элементов с индексами j и j + 1 не подходит под сортировку (в нашем
            # случае — элемент с индексом j + 1 меньше, чем элемент с индексом j), тогда меняем их местами
            if alist[j + 1] < alist[j]:
                alist[j], alist[j + 1] = alist[j + 1], alist[j]
                no_swap = False
        if no_swap:  # Если на одном из проходов по внутреннему циклу у нас нет перестановок, значит список
            # отсортирован и можно выходить из цикла
            return


# Шелла
def shell(data):
    inc = len(data) // 2
    while inc:
        for i, el in enumerate(data):  # закидывает текущий элемент в буфер и сравнивает его с левой частью массива.
            # Если находит бо́льшие элементы слева, то сдвигает их вправо, освобождая место для вставки. Но при этом
            # берёт не всю левую часть, а только некоторую группу элементов из неё, где элементы разнесены друг от
            # друга на некоторое расстояние. Такая система позволяет быстро вставлять элементы примерно в ту область
            # массива, где они должны находиться
            while i >= inc and data[i - inc] > el:
                data[i] = data[i - inc]
                i -= inc
            data[i] = el
        inc = 1 if inc == 2 else int(inc * 5.0 / 11)
    return data  # С каждой итерацией основного цикла это расстояние постепенно уменьшается и когда оно становится
    # равным единице, то сортировка Шелла в этот момент превращается в классическую сортировку простыми вставками,
    # которой дали на обработку почти отсортированный массив. А почти отсортированный массив сортировка вставками в
    # полностью отсортированный преобразует быстро


# Турнирная
def heapsort(alist):  # принимает на вход список
    build_max_heap(alist)
    for i in range(len(alist) - 1, 0, -1):  # пока размер пирамиды не станет 0 и весь список не отсортируется
        alist[0], alist[i] = alist[i], alist[0]  # Поменяем местами первый и последний элемент пирамиды
        max_heapify(alist, index=0, size=i)
# учитывая что новая пирамида имеет размер на единицу меньше. Установим index=0 для удовлетворения параметрам пирамиды


def parent(i):  # принимает index и возвращает индекс родителя
    return (i - 1) // 2


def left(i):  # принимает index и возвращает индекс левого дочернего элемента
    return 2 * i + 1


def right(i):  # принимает index и возвращает индекс правого дочернего элемента
    return 2 * i + 2


def build_max_heap(alist):  # для представления листа в виде пирамиды
    length = len(alist)  # принимает список аргументов и переставляет их в соостветсвии с max heap
    start = parent(length - 1)
    while start >= 0:
        max_heapify(alist, index=start, size=length)
        start = start - 1  # вызывает max_heapify на каждом родительском ноде и проходит до вершины


def max_heapify(alist, index, size):  # принимает индекс и изменяет структуру пирамиды на ноде и снизу от индекса
    # так, чтобы удовлетворять правилам пирамиды
    le = left(index)
    r = right(index)
    if le < size and alist[le] > alist[index]:
        largest = le
    else:
        largest = index
    if r < size and alist[r] > alist[largest]:
        largest = r
    if largest != index:
        alist[largest], alist[index] = alist[index], alist[largest]
        max_heapify(alist, largest, size)


# Быстрая сортировка
def quicksort(alist, start=-1, end=-1):
    start = 0 if start == -1 else start
    end = len(alist) if end == -1 else end
    if end - start > 1:  # Если end-start не больше 1, выходим
        p = partition(alist, start, end)  # индекс опорного элемента
        quicksort(alist, start, p)
        quicksort(alist, p + 1, end)


def partition(alist, start, end):  # использует схему разбиения Хоара
    pivot = alist[start]
    i = start + 1
    j = end - 1

    while True:
        while i <= j and alist[i] <= pivot:
            i = i + 1
        while i <= j and alist[j] >= pivot:
            j = j - 1

        if i <= j:
            alist[i], alist[j] = alist[j], alist[i]
        else:
            alist[start], alist[j] = alist[j], alist[start]
            return j


m = int(input("Введите число столбцов: ") or "50")
n = int(input("Введите число строк: ") or "50")
min_lim = int(input("Введите минимальное значение переменной: ") or "-250")
max_lim = int(input("Введите максимальное значение переменной: ") or "1012")
matr = [[random.randint(min_lim, max_lim) for j in range(m)] for i in range(n)]
print("Сформированные строки матрицы: ", matr)

test_sort(copy.deepcopy(matr), list.sort, "Стандартная:")
# .sort() производится по умолчанию лексикографически —
# проще говоря, в алфавитном порядке, а также от меньшего значения к большему
test_sort(copy.deepcopy(matr), selection_sort, "Выбором:")
test_sort(copy.deepcopy(matr), insertion_sort, "Вставкой:")
test_sort(copy.deepcopy(matr), bubble_sort, "Обменом:")
test_sort(copy.deepcopy(matr), shell, "Шелла:")
test_sort(copy.deepcopy(matr), quicksort, "Быстрая:")
test_sort(copy.deepcopy(matr), heapsort, "Пирамидальная или Турнирная:")
```
